```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcdefaults()
plt.rcParams
%matplotlib inline
plt.rcParams["figure.dpi"]=150
plt.rcParams["figure.facecolor"]="white"


data1 = np.loadtxt('../data/pp_gen/Ti_example/ld1.dlog')
data2 = np.loadtxt('../data/pp_gen/Ti_example/ld1ps.dlog')

# Extract columns: x = column 1, y = columns 2, 3, 4
x1 = data1[:, 0]
y1_2 = data1[:, 1]
y1_3 = data1[:, 2]
y1_4 = data1[:, 3]

x2 = data2[:, 0]
y2_2 = data2[:, 1]
y2_3 = data2[:, 2]
y2_4 = data2[:, 3]

# Plot data from ld1.dlog with line styles corresponding to lt 1, 2, 3
plt.plot(x1, y1_2, label='4s AE', linestyle='-', color='blue')   # lt 1
plt.plot(x1, y1_3, label='3d AE', linestyle='-', color='red')   # lt 2
plt.plot(x1, y1_4, label='4p AE', linestyle='-', color='green') # lt 3

# Plot data from ld1ps.dlog es corresponding to lt 1, 2, 3
plt.plot(x2, y2_2, label='4s PS', linestyle='-.', color='blue')   # lt 1
plt.plot(x2, y2_3, label='3d PS', linestyle='-.', color='red')   # lt 2
plt.plot(x2, y2_4, label='4p PS', linestyle='-.', color='green') # lt 3

# Set the x-axis and y-axis ranges
plt.xlim([-2, 2])
plt.ylim([-20, 20])

# Add labels and title
plt.xlabel('energy [Ry]')
plt.ylabel('Log derivative []')
plt.title('Log derivatives NC PP')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Load the data from the file
data = np.loadtxt('../data/pp_gen/Ti_example/ld1ps.wfc')

# Extract columns: x = column 1, y = columns 2, 3, 4, 5, 6, 7
x = data[:, 0]
y2 = data[:, 1]
y3 = data[:, 2]
y4 = data[:, 3]
y5 = data[:, 4]
y6 = data[:, 5]
y7 = data[:, 6]

# Plot each curve with different line types
plt.plot(x, y2, label='ps 4s', linestyle='-', color='blue')   # lt 1
plt.plot(x, y3, label='ps 3d', linestyle='-', color='green') # lt 3
plt.plot(x, y4, label='ps 4p', linestyle='-', color='red')   # lt 2
plt.plot(x, y5, label='ae 4s ', linestyle='-.', color='blue')   # lt 1 with line
plt.plot(x, y6, label='ae 3d', linestyle='-.', color='green') # lt 3 with line
plt.plot(x, y7, label='ae 4p', linestyle='-.', color='red')   # lt 2 with line

# Set the x-axis range
plt.xlim([0, 5])

# Add labels and title
plt.xlabel('distance [exponential grid]')
plt.ylabel(r'$\Phi$')
plt.title('AE and PS orbitals NC PP')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
